```sql
CREATE TABLE Cidade (
  Nome_Cidade VARCHAR(100) NOT NULL,
  Sigla_Estado VARCHAR(2) NOT NULL,
  PRIMARY KEY(Nome_Cidade)
);

CREATE TABLE Pessoa (
  Nome VARCHAR(100) NOT NULL,
  Nome_da_Mae VARCHAR(100),
  Ano_de_Nascimento INTEGER NOT NULL,
  Cidade_Natal VARCHAR(100) NOT NULL,
  
  PRIMARY KEY(Nome),
  FOREIGN KEY(Cidade_Natal)
    REFERENCES Cidade(Nome_Cidade)
      ON DELETE NO ACTION
      ON UPDATE NO ACTION,  
);
```

## Selecionando todas as pessoas que possuem primos, usando somente ALIAS  (Jeito que eu fiz antes da resposta do professor)

```sql
SELECT DISTINCT P1.nome
    From Pessoa P1, Pessoa P2, Pessoa P3
    WHERE P2.Nome = P1.Nome_da_Mae AND P3.nome = P2.Nome_da_Mae;
```

## Selecionando todas as pessoas que possuem primos, usando somente ALIAS (Depois da resposta do professor)

```sql
SELECT DISTINCT P1.nome
FROM Pessoa P1, Pessoa P2, Pessoa P3, Pessoa P4
WHERE P1.Nome_da_Mae = P2.nome AND
      P3.Nome_da_Mae = P4.nome AND
      P2.Nome_da_Mae = P4.Nome_da_Mae AND
      P2.nome <> P4.nome
```

## Selecionando todas as pessoas que possuem primos, usando dois IEWS e um ALIAS

```sql
DROP TABLE IF EXISTS Tab01;
DROP TABLE IF EXISTS Tab02;

CREATE VIEW Tab01 AS 
    SELECT P1.nome, P2.Nome_da_Mae
    FROM Pessoa P1, Pessoa P2
    WHERE P1.Nome_da_Mae = P2.Nome;
    
CREATE VIEW Tab02 AS 
    SELECT P1.nome, P2.Nome_da_Mae
    FROM Pessoa P1, Pessoa P2
    WHERE P1.Nome_da_Mae = P2.Nome;
    
SELECT DISTINCT P1.nome
    FROM Tab01 P1, Tab02 P2
    WHERE P1.Nome_da_Mae = P2.Nome_da_Mae;
```
